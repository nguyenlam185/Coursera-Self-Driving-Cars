```python
import sympy as sp
from sympy import *
from sympy.stats import Normal
import numpy as np
```

```python
x, y, theta, T, v_k, om_k, w_k, Q = symbols('x y theta_k-1 T v_k om_k w_k Q')
```

```python
sp.init_printing(use_latex=True)
```

```python
X_k = Matrix([[x], [y], [theta]]) + T * Matrix([[cos(theta), 0], [sin(theta), 0], [0, 1]]) @ (Matrix([[v_k], [om_k]]))
state = Matrix([x, y, theta])
X_k
```

```python
X_k.jacobian(state)
```

```python
xl, yl, d, y_l_k = symbols('x_l, y_l, d, y_k^l')
R11, R12, R21, R22, n_l_k= symbols('R_1^1, R_1^2, R_2^1, R_2^2, n_l_k')
n_l_k = Matrix([[R11, R12], [R21, R22]])
```

```python
y_l_k = Matrix([[sqrt((xl - x - d*cos(theta))**2 + (yl - y - d*sin(theta))**2) + n_l_k[0]], [atan2(yl - y - d * sin(theta), xl - x - d*cos(theta)) - theta + n_l_k[1]]])
y_l_k.jacobian(state)
```

```python
state_n = Matrix([n_l_k[0], n_l_k[3]])
y_l_k.jacobian(state_n)
```
