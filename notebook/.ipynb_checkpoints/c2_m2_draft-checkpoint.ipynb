```python
import sympy as sp
from sympy import *
from sympy.stats import Normal
import numpy as np
```

```python
x, y, theta, T, v_k, om_k, w_k, Q = symbols('x y theta_k-1 T v_k om_k w_k Q')
```

```python
sp.init_printing(use_latex=True)
```

```python
X_k = Matrix([[x], [y], [theta]]) + T * Matrix([[cos(theta), 0], [sin(theta), 0], [0, 1]]) @ (Matrix([[v_k], [om_k]]))
state = Matrix([x, y, theta])
X_k
```

```python
X_k.jacobian(state)
```

```python
xl, yl, d, y_l_k, xk, yk, theta_k = symbols('x_l, y_l, d, y_k^l, x_k, y_k, theta_k')
R11, R12, R21, R22 = symbols('R_1^1, R_1^2, R_2^1, R_2^2')
r_l_k = Normal('r_k^l', m1, R1)
phi_l_k = Normal('phi_k^l', m2, R2)
n_l_k = Normal('n_k^l', [0, 0], [[R11, R12], [R21, R22]])
```

```python
y_l_k = Matrix([[sqrt((xl - xk - d*cos(theta_k))**2 + (yl - yk - d*sin(theta_k))**2) + n_l_k[0]], [atan2(yl - yk - d * sin(theta_k), xl - xk - d*cos(theta_k)) - theta_k + n_l_k[1]]])
state_y = Matrix([xk, yk, theta_k])
y_l_k.jacobian(state_y)
```

```python
state_n = Matrix([n_l_k[0], n_l_k[1]])
y_l_k.jacobian(state_n)
```
